```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r'C:\Users\hp\OneDrive\Desktop\Purchase.csv')
df['purch_date'] = df['purch_date'].astype(str).str.strip()
df['purch_date'] = pd.to_datetime(df['purch_date'], dayfirst=True, errors='coerce')
df = df.dropna(subset=['purch_date'])
daily_paid = df.groupby('purch_date')['paid'].sum()
plt.figure(figsize=(10,6))
daily_paid.plot(kind='bar',color='skyblue')
plt.title('Total Paid per Day')
plt.xlabel('Purchase Date')
plt.ylabel('Total Paid')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
daily_paid = df.groupby('purch_date')['paid'].sum()
plt.figure(figsize=(10, 5))
plt.plot(daily_paid.index, daily_paid.values, marker='o', linestyle='-', color='blue')
plt.title('Total Paid Over Time')
plt.xlabel('Purchase Date')
plt.ylabel('Total Paid')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_customers = df.groupby('customer_num')['paid'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top_customers.plot(kind='bar', color='green')
plt.title('Top 10 Customers by Paid Amount')
plt.xlabel('Customer Number')
plt.ylabel('Total Paid')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
top_products = df.groupby('product_num')['paid'].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(7, 7))
plt.pie(top_products, labels=top_products.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Products by Paid Amount')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['paid'], bins=15, color='orange', edgecolor='black')
plt.title('Distribution of Paid Amounts')
plt.xlabel('Amount Paid')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['amount'], df['paid'], alpha=0.6, color='purple')
plt.title('Amount vs Paid')
plt.xlabel('Amount')
plt.ylabel('Paid')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r'C:\Users\hp\OneDrive\Desktop\Purchase.csv')
df['purch_date'] = pd.to_datetime(df['purch_date'], dayfirst=True, errors='coerce')
df = df.dropna(subset=['purch_date'])
```

```python
df['month'] = df['purch_date'].dt.to_period('M')
monthly_paid = df.groupby('month')['paid'].sum()
plt.figure(figsize=(10, 6))
monthly_paid.plot(marker='o', color='teal')
plt.title('Monthly Total Paid')
plt.xlabel('Month')
plt.ylabel('Total Paid')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['week'] = df['purch_date'].dt.to_period('W')
weekly_paid = df.groupby('week')['paid'].sum()
plt.figure(figsize=(10, 6))
weekly_paid.plot(kind='bar', color='coral')
plt.title('Weekly Total Paid')
plt.xlabel('Week')
plt.ylabel('Total Paid')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
daily_paid = df.groupby('purch_date')['paid'].sum()
cumulative_paid = daily_paid.cumsum()
plt.figure(figsize=(10, 6))
plt.plot(cumulative_paid.index, cumulative_paid.values, marker='.', linestyle='-', color='green')
plt.title('Cumulative Paid Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Paid')
plt.grid(True)
plt.tight_layout()
plt.show()
```
